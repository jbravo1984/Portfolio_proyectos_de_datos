```python
#Evolución de la esperanza de vida en Chile a nivel nacional

Chile = df.iloc[332, 5:]
```

```python
#Gráfico único de la esperanza de vida en Chile

import matplotlib.pyplot as plt

Chile.plot()
plt.xlabel('Años')
plt.ylabel('Edad de esperanza de vida')
plt.title('Evolución de la esperanza de vida en Chile')
```

```python
#Evolución de la esperanza de vida en Perú a nivel nacional

Peru = df.iloc[1436, 5:]
```

```python
#Concatenación de ambas series de datos

import pandas as pd

Comparacion = pd.concat([Chile, Peru], axis='columns')
```

```python
#Cambio de nombre en las columnas

Comparacion['Chile'] = Comparacion[332]
Comparacion['Perú'] = Comparacion[1436]
```

```python
Comparacion1 = Comparacion.drop(332, axis='columns')
```

```python
Comparacion2 = Comparacion1.drop(1436, axis='columns')
```

```python
Comparacion2.plot()
plt.xlabel('Años')
plt.ylabel('Esperanza de vida')
plt.title('Comparativa de la esperanza de vida de Chile y Perú')
```
